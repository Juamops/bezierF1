```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from bCurve import Bezier, Point, Spline
import numpy as np
from PIL import Image
```

```python
def hypot(x, y):
    return (x**2 + y**2)**0.5

def distance(p1, p2):
    return (((p1.x - p2.x) ** 2) + ((p1.y - p2.y) ** 2)) ** 0.5
```

```python
track_spline = Spline([
    Bezier(Point(251, 626), Point(263, 646), Point(286.7868127133783, 658.0197535168365), Point(308, 660)),
    Bezier(Point(308, 660), Point(329.2131872866217, 661.9802464831635), Point(358.48498177436346, 655.1220565995621), Point(386, 650)),
    Bezier(Point(386, 650), Point(413.51501822563654, 644.8779434004379), Point(435.34014682305747, 639.276416871691), Point(461, 627)),
    Bezier(Point(461, 627), Point(486.65985317694253, 614.723583128309), Point(517.8956587742977, 598.0903132675311), Point(537, 576)),
    Bezier(Point(537, 576), Point(556.1043412257023, 553.9096867324689), Point(562.0, 543.0), Point(571, 508)),
    Bezier(Point(571, 508), Point(580.0, 473.0), Point(559.0, 410.0), Point(563, 380)),
    Bezier(Point(563, 380), Point(567.0, 350.0), Point(600.0, 305.0), Point(610, 280)),
    Bezier(Point(610, 280), Point(620.0, 255.0), Point(623.0, 224.0), Point(635, 193)),
    Bezier(Point(635, 193), Point(647.0, 162.0), Point(662.0033989403206, 144.3387909605964), Point(699, 115)),
    Bezier(Point(699, 115), Point(735.9966010596794, 85.66120903940359), Point(771, 64), Point(793, 47)),
])
track_spline.make_continuous()

in_to_p = 404/16583
m_to_in = 39.3701
kmh_to_ms = 1/3.6
kmh_to_p = m_to_in * in_to_p * kmh_to_ms
```

```python
track = np.asarray(Image.open('track.png'))
plt.imshow(track)
track_x, track_y = track_spline.get_points(0, 1, 500)
control_line_x, control_line_y = track_spline.get_control_lines()
control_point_x, control_point_y = track_spline.get_control_points()
plt.plot(track_x, track_y, lw=5, color='g')

for x_end, y_end in zip(control_line_x, control_line_y):
    plt.plot(x_end, y_end, color='r')

plt.scatter(control_point_x, control_point_y, color='b', s=7, zorder=10)

plt.axis('equal')
plt.xlim([200, 900])
plt.ylim([700, 0])
plt.show()
```

```python
x, y = track_spline.get_points(0, 1, 100)
x_d, y_d = track_spline.get_first_derivatives(0, 1, 100)
x_d2, y_d2 = track_spline.get_second_derivatives(0, 1, 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, lw=3, zorder=0)

# animated_track, = ax.plot([], [], lw=3)
v_scale = 0.5
animated_point, = ax.plot(x[0], y[0], color='r', marker='o', markersize=6, zorder=15)
animated_velocity = ax.arrow(x[0], y[0], x_d[0]*v_scale, y_d[0]*v_scale, color='b', width=2, head_width=10, zorder=10)
animated_acceleration = ax.arrow(x[0], y[0], x_d2[0]*v_scale, y_d2[0]*v_scale, color='g', width=2, head_width=10, zorder=10)


def update(frame):
    # animated_plot.set_linewidth(3)
    # animated_track.set_data(x[:frame], y[:frame])
    animated_velocity.set_data(x=x[frame], y=y[frame], dx=x_d[frame]*v_scale, dy=y_d[frame]*v_scale)
    animated_acceleration.set_data(x=x[frame], y=y[frame], dx=x_d2[frame]*v_scale, dy=y_d2[frame]*v_scale)
    animated_point.set_data([x[frame]], [y[frame]])

    return animated_point,


animation = FuncAnimation(
        fig=fig,
        func=update,
        frames=len(x),
        interval=1
)
ax.axis('equal')
ax.set_xlim([200, 900])
ax.set_ylim([700, 0])
animation.save('report_p_traverse.gif')
```

```python
fig, ax = plt.subplots(1, 2)

# animated_track, = ax.plot([], [], lw=3)
v_scale = 0.5
animated_velocity_v = ax[0].arrow(0, 0, x_d[0], y_d[0], color='b', head_width=6, zorder=10)
animated_velocity_p, = ax[0].plot([], [], color='b', zorder=10)

animated_acceleration_v = ax[1].arrow(0, 0, x_d2[0], y_d2[0], color='g', head_width=6, zorder=10)
animated_acceleration_p, = ax[1].plot([], [], color='g', zorder=10)


def update(frame):
    # animated_plot.set_linewidth(3)
    # animated_track.set_data(x[:frame], y[:frame])
    animated_velocity_v.set_data(x=0, y=0, dx=x_d[frame], dy=y_d[frame])
    animated_velocity_p.set_data(x_d[:frame], y_d[:frame])
    
    animated_acceleration_v.set_data(x=0, y=0, dx=x_d2[frame], dy=y_d2[frame])
    animated_acceleration_p.set_data(x_d2[:frame], y_d2[:frame])


animation = FuncAnimation(
        fig=fig,
        func=update,
        frames=len(x),
        interval=1
)
ax[0].axis('equal')
ax[0].set_aspect('equal')
ax[0].set_xlim([-300, 300])
ax[0].set_ylim([-300, 300])

ax[1].axis('equal')
ax[1].set_aspect('equal')
ax[1].set_xlim([-300, 300])
ax[1].set_ylim([-300, 300])
animation.save('report_p_stats.gif')
```

```python
x, y = track_spline.get_points(0, 1, 100)
const_t, const_time, d = track_spline.traverse_const_vel(0, 1, 220*kmh_to_p, timestep=0.1)
vel_x, vel_y, acc_x, acc_y = track_spline.get_movement_vectors(const_t, const_time)
```

```python
danger_x = [track_spline.get_point_norm(t).x + vel_x[i]*1.5 if track_spline.get_max_vel(t, 1.7) < 2/3.6 else 0 for i, t in enumerate(const_t)]
danger_y = [track_spline.get_point_norm(t).y + vel_y[i]*1.5 if track_spline.get_max_vel(t, 1.7) < 2/3.6 else 0 for i, t in enumerate(const_t)]
```

```python
plt.plot(x, y, lw=3, zorder=0)
plt.scatter(danger_x, danger_y, color='r')


plt.axis('equal')
plt.xlim([200, 900])
plt.ylim([700, 0])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, lw=3, zorder=0)

# animated_track, = ax.plot([], [], lw=3)
v_scale = 0.3
animated_point, = ax.plot(x[0], y[0], color='r', marker='o', markersize=6, zorder=15)
animated_velocity = ax.arrow(x[0], y[0], vel_x[0]*v_scale, vel_y[0]*v_scale, color='b', width=2, head_width=10, zorder=10)
animated_acceleration = ax.arrow(x[0], y[0], acc_x[0]*v_scale, acc_y[0]*v_scale, color='g', width=2, head_width=10, zorder=10)


def update(frame):
    # animated_plot.set_linewidth(3)
    # animated_track.set_data(x[:frame], y[:frame])
    point = track_spline.get_point_norm(const_t[frame])
    animated_velocity.set_data(x=point.x, y=point.y, dx=vel_x[frame]*v_scale, dy=vel_y[frame]*v_scale)
    animated_acceleration.set_data(x=point.x, y=point.y, dx=acc_x[frame]*v_scale, dy=acc_y[frame]*v_scale)
    animated_point.set_data([point.x], [point.y])

    return animated_point,


animation = FuncAnimation(
        fig=fig,
        func=update,
        frames=len(const_t),
        interval=1
)
ax.axis('equal')
ax.set_xlim([200, 900])
ax.set_ylim([700, 0])
animation.save('report_c_traverse.gif')
```
